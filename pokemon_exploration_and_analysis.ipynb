## How does height and weight of a Pokemon correlate with its various base stats?

```python
height_weight_and_base_stats = pokemon[['height_m', 'weight_kg', 'hp', 'attack', 'defense', 'speed', 'sp_attack', 'sp_defense']]
height_weight_and_base_stats
```

```python
sns.pairplot(data=height_weight_and_base_stats, x_vars=['hp', 'attack', 'defense', 'speed', 'sp_attack', 'sp_defense'], y_vars=['height_m', 'weight_kg'])
```

```python
ax = sns.heatmap(height_weight_and_base_stats.corr(), annot=True, cmap="Blues")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
